```python
from matplotlib import pyplot as plt 
from scipy import stats
import numpy as np  
import pandas as pd
import csv

b = 6.17*(10**-5)
P = 706
vis = 18.35*(10**-6)
dens = 890
g = 9.806

def load_data(filename, col1, col2):
    rise = []; fall = []
    with open(filename) as numbers:
        numbers_data = csv.reader(numbers, delimiter=",")
        next(numbers_data) #skip the header
        for row in numbers_data:
            if row[col1] == '' or row[col2] == '':
                pass
            else:
                rise.append(float(row[col1])); fall.append(float(row[col2]))
        return rise, fall

velo_rise_err = []; velo_fall_err = []
counter = 1; inc = 1; time_data = {}
while counter <= 17:
    fight = load_data("PHYS 295 Lab 5 (Millikan Oil Drop).csv", inc - 1, inc)
    time_data[f"Droplet {counter}"] = [round(np.mean(fight[0]), 2), 
                                       round(np.mean(fight[1]), 2), 
                                       round(stats.sem(fight[0]), 2),   
                                       round(stats.sem(fight[1]), 2)]
    temp1 = []; temp2 = []
    for arr in fight[0]:
        temp1.append(0.0006/arr)
    for arr in fight[1]:
        temp2.append(0.0006/arr)
    velo_rise_err.append(round(stats.sem(temp1), 9)); velo_fall_err.append(round(stats.sem(temp2), 10)) 
    counter += 1; inc += 2
    
#print(velo_rise_err, "\n")
#print(velo_fall_err, "\n")

velo_data = {}
rise_err_data = []; fall_err_data = []
for value in time_data:
    temp1 = []; temp2 = []
    rise_speed = round(0.0006/time_data[value][0], 7); fall_speed = round(0.0006/time_data[value][1], 7)
    temp1.append(rise_speed); temp1.append(fall_speed)
    rise_err_data.append(time_data[value][2]); fall_err_data.append(time_data[value][3]) 
    velo_data[value] = temp1

v_rise = []; v_fall = []
for value in velo_data:
    print(f"{value}:")
    print(time_data[value][0], time_data[value][1])
    print(f"The rise velocity is {velo_data[value][0]} m/s")
    print(f"The fall velocity is {velo_data[value][1]} m/s\n")
    v_rise.append(velo_data[value][0])
    v_fall.append(velo_data[value][1])
```

```python
k_value = []
q_value1 = []; q_value2 = []; n_value = []; rd = []; QE = [];
for value in velo_data:
    radius = ((-b)/(2*P)) + np.sqrt(((b**2)/(4*(P**2))) + ((9*vis*velo_data[value][1])/(2*dens*g)))
    rd.append(radius)
    k = (6*np.pi*radius*vis)/(1 + (b/(P*radius)))
    if value == "Droplet 1":
        print(f"The k value for Droplet 1 is {k}")
    qE = k*(velo_data[value][0] + velo_data[value][1])
    QE.append(qE)
    E = 400/0.005693
    m = (4/3)*np.pi*dens*(radius**3)
    q1 = qE/E
    q2 = (1 + (velo_data[value][0]/velo_data[value][1]))*(g/E)*m
    n = q1/((5.598*(10**-19))/3)
    #print(f"The value of the electric field is {E}")
    print(f"The radius of {value} is equal to {radius}")
    #print(f"The qE value of {value} is equal to {qE}\n")
    k_value.append(round(k, 13))
    q_value1.append(round(q1, 21))
    q_value2.append(round(q2, 21)/3)
    n_value.append(round(n, 2))

FB = (4/3)*np.pi*((rd[0])**3)*1.293*g
print(f"\nThe buoyant force for Droplet 1 is {FB}")
print(f"The qE value for Droplet 1 is {QE[0]}")
```

```python
q_err = []
counter = 0
for value in velo_data:
    q = q_value1[counter]
    vre = velo_rise_err[counter]
    vr = velo_data[value][0]
    vfe = velo_fall_err[counter]
    vf = velo_data[value][1]
    charge_error = q*np.sqrt(((vre/(vr + vf))**2) + (((1/(2*vf)) + (1/(vr + vf)))**2)*(vfe**2))
    q_err.append(charge_error)
    counter += 1
```

```python
table = {"Droplet Number": time_data.keys(),
         "k value": k_value,
         "Rise V": v_rise,
         "Rise Time Err": rise_err_data,
         "Rise V Err": velo_rise_err,
         "Fall V": v_fall, 
         "Fall Time Err": fall_err_data,
         "Fall V Err": velo_fall_err,
         "Charge M1(q)": q_value1, 
         "Charge M2(q)": q_value2, 
         "Charge Err (δq)": q_err, 
         "n": n_value}
 
show_table = pd.DataFrame(data=table)
show_table
```
